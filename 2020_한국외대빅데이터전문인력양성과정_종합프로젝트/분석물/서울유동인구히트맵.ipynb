```python
# matplotlib 한글설정
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

f_path = "C:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=f_path).get_name()
rc('font', family=font_name)
```

```python
seoul_people = pd.read_csv("서울시유동인구.csv",
                           encoding = "UTF-8")
seoul_people.index = seoul_people["행 레이블"]
del(seoul_people["행 레이블"])
seoul_people.head()
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(seoul_people, annot=True, fmt='f', linewidths=.5)
plt.title("서울시 시간대별 평균 유동인구")
plt.show()
```
