```python
star = pd.read_excel("전국스타벅스DT매장주소.xlsx")
star.head()
```

## starbuck 주소별 빈도표 만들기

```python
star_count = star.groupby("jsonid").count()
star_count.describe()
```

## 전국 스타벅스 빈도 지도 시각화

```python
map = folium.Map(location=[37.56, 126.98],
                 zoom_start=11)

geo_path = 'full_skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))

map.choropleth(
geo_data=geo_str,
data=star_count,
columns=[star_count.index, '지점명'],
key_on='feature.id',
fill_color='PuRd',
threshold_scale = [0, 1, 3, 6, 9, 76, 100])

map
```

## 전국 스타벅스 "DT" 매장 빈도표 만들기

```python
star_DT = pd.read_excel("전국스타벅스DT매장주소.xlsx",
                       sheet_name = "스타벅스DT매장주소(20191029)")
star_DT.head()
```

```python
star_DT_count = star_DT.groupby("jsonid").count()
star_DT_count.describe()
```

## 지도 시각화

```python
map = folium.Map(location=[37.56, 126.98],
                 zoom_start=11)

geo_path = 'full_skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))

map.choropleth(
geo_data=geo_str,
data=star_DT_count,
columns=[star_DT_count.index, '지점명'],
key_on='feature.id',
fill_color='PuRd',
threshold_scale = [0, 1, 2, 3, 8, 10])

map
```
